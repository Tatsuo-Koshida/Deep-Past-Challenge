```python
# ソース：https://www.kaggle.com/competitions/deep-past-initiative-machine-translation/discussion/678899#3415774
pub_pd = pd.read_csv("/Users/koshidatatsuo/python/kaggle/Deep_Past_Challenge/data/kaggle/deep-past-initiative-machine-translation/published_texts.csv")

train_df = pd.read_csv("/Users/koshidatatsuo/python/kaggle/Deep_Past_Challenge/data/kaggle/deep-past-initiative-machine-translation/train.csv")

akt_pd = pub_pd[(pub_pd['oare_id'].isin(train_df["oare_id"].unique()))][["oare_id", "label", "excavation_no", "transliteration"]]
akt_pd["pdf"] = akt_pd["label"].str.extract(
    r"\bAKT\s+(\d+[a-zA-Z]?)",
    expand=False
)
```
